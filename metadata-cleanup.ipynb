```python
import pandas as pd
import numpy as np
```

```python
cols_of_interest = ['sample_id', 'barcode', 'username', 'station_specific', 'line',
                   'location', 'geolocation', 'ground_level',
                   'sampling_place', 'sampling_place_other',
                   'sampling_place_specifics', 'sampling_place_id',
                   'surface_material', 'surface_material_other', 'picture', 'note']

geometa = pd.read_excel('data/metadata_withgeo.xlsx', usecols=cols_of_interest)
#reorder
geometa = geometa[cols_of_interest]
geometa['barcode'] = geometa['barcode'].astype(str)
```

```python
meta = pd.read_csv('data/complete_metadata.csv')

meta = meta[
    (meta['city'] == 'kyiv') &
    (meta['station'].notna()) &
    (meta['station'] != 'busstop_20')
]

cols_of_interest = ['barcode', 'uuid', 'sl_name',
                    'surface_material', 'station', 'surface', 'elevation',
                    'num_reads', 'hudson_alpha_flowcell', 'index_sequence']

meta = meta[cols_of_interest]
meta['barcode'] = meta['barcode'].astype(int).astype(str)
```

```python
merged = meta.merge(geometa, on = 'barcode', suffixes=('', '_geo'))
```

```python
merged['surface'] = np.where(merged['surface'] == 'other', merged['sampling_place_other'], merged['surface'])
```

```python
merged['Станція'] = merged['station'].replace({'maidan_nezalezhnosti' : 'Mайдан Незалежності',
                           'minska' : 'Мінська',
                           'petrivka' : 'Почайна',
                           'heroiv_dnipra' : 'Героїв Дніпра',
                           'kontraktova_ploshcha' : 'Контрактова площа'})

merged['Матеріал'] = merged['surface_material'].replace({'metal' : 'метал',
                                                         'wood' : 'дерево',
                                                         'granite' : 'граніт',
                                                         'plastic' : 'пластик',
                                                         'rubber' : 'гума',
                                                         'painted_metal' : 'фарба'})

merged['Поверхня'] = merged['surface'].replace({
    'jetton' : 'жетон',
    'bench' : 'лавка',
    'entrance_door' : 'двері',
    'floor' : 'підлога',
    'horiz_pole' : 'горизонтальний поручень',
    'ticket_machine' : 'квитковий автомат',
    'escalator_handrail' : 'поручень ескалатора',
    'handrail' : 'вертикальний поручень',
    'seat' : 'сидіння',
    'tourniquet' : 'турнікет'
})
```

```python
(merged
 .drop(['ground_level', 'sampling_place', 'sampling_place_other',
        'surface_material_other', 'location', 'note'], axis = 1)
 .to_csv('data/clean_metadata.csv', index = False))
```
